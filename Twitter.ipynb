```python
location = []

for index, row in tweets.iterrows():
    text = re.sub('[^A-Za-z ]+', '', row['location'])
    location.append(text.strip())
    
```

```python
print(len(location), "Locations")
```

```python
tweets['location'] = location
tweets['location'] = list(np.where(tweets['location'] == " " , 'Location Blocked', tweets['location']))
tweets['location'] = list(np.where(tweets['location'] == "" , 'Location Blocked', tweets['location']))
tweets['location'] = list(np.where(len(tweets['location']) <= 1 , 'Location Blocked', tweets['location']))
```

# Sentiment Analysis

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

%matplotlib inline

from IPython.core.display import HTML
```

```python
#removedURL = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", reviewSentence)
```

```python
sid = SentimentIntensityAnalyzer()

fullText = ""

spChar = "'"

sentimentRes = ""

se = [] #Empty array

twit = []

for index, row in Replied_tweets.iterrows():
    
    reviewSentence = str(row["tweet"]) #selecting comments from each row
    
    pre_text = re.sub(r'@[A-Za-z0-9]+','', reviewSentence)
    
    pro_text = re.sub('https?://[A-Za-z0-9./]+','', pre_text)
    
    reviewSentence = re.sub('[^ a-zA-Z]', '', pro_text) #Removing special characters form the sentence
    
    reviewSentence = reviewSentence.strip().lower()
    
    ss = sid.polarity_scores(reviewSentence)
    
    #defining the result of each description
    sentiment = ""

    if ss['compound'] == 0:
        sentiment = "Neutral"
    elif ss['compound'] > 0:
        sentiment = "Positve"
    else:
        sentiment = "Negative"
    
    #adding values into array
    se.append(sentiment)
    twit.append(reviewSentence)

print(ss)
```

```python
#For getting sentiment output from array
sentiment_result = ""
for i in range(0,len(se)):
    sentiment_result = sentiment_result + se[i] +"','"

sentiment_result = "'"+sentiment_result[0:len(sentiment_result)-2]
#sentiment_result
```

```python
Replied_tweets['sentiment'] = se
Replied_tweets['preprocessed_twit'] = twit
```
